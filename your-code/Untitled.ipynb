```python
challenge3 = requests.get('https://api.github.com/repos/ironhack-datalabs/scavenger/git/trees/master?recursive=1')
challenge3 = challenge3.json()
```

```python
df_c3 = json_normalize(challenge3['tree'])
df_c3 = df_c3[df_c3['path'].str.contains('\d+\/\.\d+\..*', regex = True)]
lista = list(df_c3['path'])
lista.sort(key = lambda x: x[9:])
lista
r = []
for el in lista:
    read = requests.get(f'https://api.github.com/repos/ironhack-datalabs/scavenger/contents{el}')
    r.append(read)
resultado = []
for el in r:
    resultado.append(el.json())
resultado_df = pd.DataFrame(resultado)
resultado_df
```

```python
import base64

codigo = list(resultado_df['content'])
for c in codigo:
    print(base64.b64decode(c))
```
